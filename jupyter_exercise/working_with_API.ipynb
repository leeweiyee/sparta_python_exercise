```python
import numpy as np
import pandas as pd
import requests
import urllib.request
import time
import json
from bs4 import BeautifulSoup
# website datasets are stored as API (application programming interface)
# APIs are used to retrieve information from any major website
# API in json format needs to be converted to dictionaries, which can be converted to dataframes
```

```python
url = 'https://api.exchangeratesapi.io/latest'
r = requests.get(url)
print(r.text)
```

```python
r = requests.get(url,params = {'base':'USD'})
print(r.text)
```

```python
type(r.text)
# current json file consists of string datatypes which need to be converted to dictionary
```

```python
# json.loads() converts json to dictionary
json_dict = json.loads(r.text)
json_dict
```

```python
# extract just rates
rates_json = json.loads(r.text)['rates']
rates_json
```

```python
type(rates_json)
```

```python
# print base values
print(rates_json['GBP'])
gbp = float(rates_json['GBP'])
print('200USD = ',str(gbp*200),'GBP')
```

```python
# save dictionary into a variable
rates = [(rates_json)]
```

```python
# create dataframe from dictionary
rates_df = pd.DataFrame(rates)
rates_df
```

```python
rates_df1 = rates_df.transpose()
rates_df1
```
