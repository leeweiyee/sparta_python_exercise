# Data Visualisation

```python
# load pre-exisitng dataset: sns.load_dataset() 
# create histogram: plt.hist(array)
# create joint plot: sns.jointplot(array1, array2)
# create smooth histogram: sns.distplot(array)
# create regression model: sns.lmplot()
# create simple linear regression model: sns.regplot()
# create heatmap: sns.heatmap()
# create clustermap: sns.clustermap()
```

## Histograms

```python
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset1 = randn(100)
dataset1
```

```python
# histograms show the frequency on the vertical axis and the bins in the horizontal axis
# each bin has a min and max value
plt.hist(dataset1)
```

```python
dataset2 = randn(80)
plt.hist(dataset2,color='indianred') # colour can be specified
```

```python
# to show both histograms in the same historgram, norm should be set to True
# alpha is the transparency
plt.hist(dataset1, normed=True, alpha=0.5, bins=20)
plt.hist(dataset2, normed=True, color='indianred', alpha=0.5, bins=20)
```

```python
dataset3 = randn(1000)
dataset4 = randn(1000)
```

```python
sns.jointplot(dataset3, dataset4)
```

```python
sns.jointplot(dataset3, dataset4, kind='hex')
```

```python
# shows highest peak clearly
sns.distplot(dataset4, bins=30)
```

```python
sns.distplot(dataset4, bins=30,rug=True, hist=False)
```

```python
# kde controls parameters which control the graph
sns.distplot(dataset4,bins = 30,
             kde_kws = {'color':'indianred','label':'Plot for KDE'},
             hist_kws = {'color':'blue','label':'Plot for Histogram'})
```

```python
# a histogram can be created from a series as well
ser1 = pd.Series(dataset4, name='my_data')
ser1
```

```python
sns.distplot(ser1, bins=30)
```

## Regression models

```python
iris_df = sns.load_dataset('iris')
iris_df
```

```python
# regression model shows relationship between a dependent (target) and independent variable 
# regplot() performs a simple linear regression model fit and plot
sns.regplot('petal_length','petal_width',iris_df)
```

```python
# lmplot() combines regplot() and FacetGrid
# FacetGrid class helps in visualising the distribution of one variable and the relationship
# between multiple variables separately within subsets of your dataset using multiple panels
sns.lmplot('petal_length','petal_width',iris_df,hue='species')
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.lmplot('total_bill','tip',tips_df)
```

```python
tips_df['tips_pect']=100*(tips_df['tip']/tips_df['total_bill'])
tips_df
```

```python
tips_df.head()
```

```python
sns.lmplot('size','tips_pect',tips_df,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('size','tips_pect',tips_df,hue = 'day')
```

```python
sns.regplot('total_bill','tips_pect',tips_df)
```

## Heat maps and cluster maps

```python
flights_df = sns.load_dataset('flights')
flights_df.head()
```

```python
flights_df_pivot = flights_df.pivot('month','year','passengers')
flights_df_pivot
```

```python
sns.heatmap(flights_df_pivot)
```

```python
sns.heatmap(flights_df_pivot,annot=True,fmt='d')
# annot = annotation
# fmt = format; has to be 'd'
```

```python
sns.heatmap(flights_df_pivot,center=flights_df_pivot.loc['January',1960],annot=True,fmt='d')
```

```python
sns.clustermap(flights_df_pivot,annot=True,fmt='d')
```

```python
sns.clustermap(flights_df_pivot,col_cluster=False,annot=True,fmt='d')
# col_cluster = cluster at column
```
